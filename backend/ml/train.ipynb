```python
import csv
import glob
import os
import numpy as np
import torch
from torch import nn

import sys
sys.path.append(r'C:\Users\KIIT0001\Desktop\gitclones\GravNet-ML')

from backend.ml.model import StabilityNet
```

Load CSV datasets

```python
PROJECT_ROOT = r"C:\Users\KIIT0001\Desktop\gitclones\GravNet-ML"
DATA_DIR = os.path.join(PROJECT_ROOT, "data", "datasets")
CSV_FILES = glob.glob(os.path.join(DATA_DIR, "dataset_*.csv"))


X, y = [], []

for path in CSV_FILES:
    with open(path) as f:
        for r in csv.DictReader(f):
            X.append([
                float(r["n_init"]),
                float(r["steps"]),
                float(r["heuristic_stability"]),
                float(r["chaos"])
            ])
            y.append(int(r["merged"]))

X = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.float32).unsqueeze(1)

print("X shape:", X.shape)
print("y shape:", y.shape)
```

Train Model

```python
model = StabilityNet(4)
opt = torch.optim.Adam(model.parameters(), 1e-3)
loss_fn = torch.nn.BCELoss()

for e in range(100):
    opt.zero_grad()
    loss = loss_fn(model(X), y)
    loss.backward()
    opt.step()
    if e % 10 == 0:
        print(e, loss.item())
```

Save Models

```python
os.makedirs(r"C:\Users\KIIT0001\Desktop\gitclones\GravNet-ML\models", exist_ok=True)
torch.save(model.state_dict(), r"C:\Users\KIIT0001\Desktop\gitclones\GravNet-ML\models\stability.pt")
print("Saved models/stability.pt")
```
